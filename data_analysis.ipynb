```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/cleaned_data.csv')
df.head(20)
```

```python
# Get columns with int or float dtypes
num_cols = df.select_dtypes(include=['int', 'float']).columns
# Display numerical summaries
df[num_cols].describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Diagnosis labels if not already present
df['DiagnosisLabel'] = df['diagnosis'].replace({0: "No Alzheimer's", 1: "Alzheimer's"})

# Set visual style
sns.set(style="whitegrid")
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Boxplot for BMI
sns.boxplot(x='DiagnosisLabel', y='bmi', data=df, ax=axes[0], palette=['#3498db', '#e74c3c'])
axes[0].set_title('BMI by Diagnosis')
axes[0].set_xlabel('')
axes[0].set_ylabel('BMI')

# Boxplot for Alcohol Consumption
sns.boxplot(x='DiagnosisLabel', y='alcoholconsumption', data=df, ax=axes[1], palette=['#3498db', '#e74c3c'])
axes[1].set_title('Alcohol Consumption by Diagnosis')
axes[1].set_xlabel('')
axes[1].set_ylabel('Alcohol Consumption (units/week)')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Diagnosis label
df['DiagnosisLabel'] = df['diagnosis'].replace({0: "No Alzheimer's", 1: "Alzheimer's"})

# Set style
sns.set(style="whitegrid")
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Boxplots
sns.boxplot(x='DiagnosisLabel', y='physicalactivity', data=df, ax=axes[0], palette=['#3498db', '#e74c3c'])
axes[0].set_title('Physical Activity by Diagnosis')

sns.boxplot(x='DiagnosisLabel', y='dietquality', data=df, ax=axes[1], palette=['#3498db', '#e74c3c'])
axes[1].set_title('Diet Quality by Diagnosis')

sns.boxplot(x='DiagnosisLabel', y='sleepquality', data=df, ax=axes[2], palette=['#3498db', '#e74c3c'])
axes[2].set_title('Sleep Quality by Diagnosis')

for ax in axes:
    ax.set_xlabel('')
    ax.set_ylabel('Score')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Map diagnosis and behavioral problems to readable labels
df['DiagnosisLabel'] = df['diagnosis'].replace({0: "No Alzheimer's", 1: "Alzheimer's"})
df['BehavioralProblemsLabel'] = df['behavioralproblems'].replace({0: "No Behavioral Problems", 1: "Has Behavioral Problems"})

# Group and calculate percentages
behavior_counts = df.groupby(['DiagnosisLabel', 'BehavioralProblemsLabel']).size().unstack()
behavior_percent = behavior_counts.div(behavior_counts.sum(axis=1), axis=0) * 100

# Plot with blue and red colors
colors = ['#3498db', '#e74c3c']  # Blue = No Problems, Red = Has Problems

behavior_percent.plot(
    kind='bar',
    stacked=True,
    color=colors,
    figsize=(8, 6),
    edgecolor='black'
)

plt.title('Behavioral Problems by Diagnosis', fontsize=16)
plt.xlabel('Diagnosis', fontsize=14)
plt.ylabel('Percentage of Individuals', fontsize=14)
plt.xticks(rotation=0)
plt.legend(title='Behavioral Problems')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate with percentage labels
for i, (idx, row) in enumerate(behavior_percent.iterrows()):
    plt.text(i - 0.2, row[0] / 2, f"{row[0]:.1f}%", color='white', fontweight='bold')
    plt.text(i - 0.2, row[0] + row[1] / 2, f"{row[1]:.1f}%", color='white', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

box = plt.boxplot(
    [df[df['diagnosis'] == 0]['age'], df[df['diagnosis'] == 1]['age']],
    patch_artist=True,
    tick_labels=["No Alzheimer's (0)", "Alzheimer's (1)"]  # ✅ For Matplotlib 3.9+
)

# Custom colors
colors = ['#3498db', '#e74c3c']  # Blue, Red

# Set colors for box fill
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# Customize median line
for median in box['medians']:
    median.set(color='black', linewidth=2)

# Add titles and labels
plt.title('Boxplot of Age by Diagnosis Group', fontsize=16)
plt.ylabel('Age', fontsize=14)
plt.xlabel('Diagnosis', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Risk factors and display labels
risk_factors = {
    'hypertension': 'Hypertension',
    'depression': 'Depression',
    'headinjury': 'Head Injury',
    'diabetes': 'Diabetes',
    'cardiovasculardisease': 'Cardiovascular Disease',
    'familyhistoryalzheimers': 'Family History',
    'smoking': 'Smoking'
}

colors = ['#3498db', '#e74c3c']  # Blue = No Alzheimer's, Red = Alzheimer's

# Create subplots: 2x4 grid but only fill 7
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten()

# Loop and plot
for i, (col, label) in enumerate(risk_factors.items()):
    # Group and calculate percentages
    group = df.groupby([col, 'diagnosis']).size().unstack()
    percent = group.div(group.sum(axis=1), axis=0) * 100

    # Rename index
    if col == 'smoking':
        percent.index = ['Non-Smoker', 'Smoker']
    elif col == 'familyhistoryalzheimers':
        percent.index = ['No Family History', 'Has Family History']
    else:
        percent.index = ['No', 'Yes']

    # Plot
    percent.plot(kind='bar', stacked=True, color=colors, ax=axes[i], edgecolor='black')
    axes[i].set_title(f'Alzheimer\'s by {label}', fontsize=14)
    axes[i].set_ylabel('Percentage')
    axes[i].set_ylim(0, 100)
    axes[i].set_xticklabels(percent.index, rotation=0)
    axes[i].grid(axis='y', linestyle='--', alpha=0.6)
    axes[i].legend().set_visible(False)

    # Annotations
    for j, (idx, row) in enumerate(percent.iterrows()):
        axes[i].text(j - 0.15, row[0] / 2, f"{row[0]:.1f}%", color='white', fontweight='bold')
        axes[i].text(j - 0.15, row[0] + row[1] / 2, f"{row[1]:.1f}%", color='white', fontweight='bold')

# Remove the empty 8th plot
fig.delaxes(axes[-1])

# Add legend and title
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, ["No Alzheimer's", "Alzheimer's"], loc='lower center', ncol=2, fontsize=12)
plt.suptitle("Alzheimer's Diagnosis Breakdown by Risk Factors", fontsize=18, y=1.02)
plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Create new education group
def group_education(level):
    if level == 0:
        return 'No Formal Education'
    else:
        return 'Some Formal Education'

df['EducationGroup'] = df['educationlevel'].apply(group_education)

# Step 2: Count diagnosis within each group
grouped_counts = df.groupby(['EducationGroup', 'diagnosis']).size().unstack()

# Step 3: Convert counts to percentages
grouped_percent = grouped_counts.div(grouped_counts.sum(axis=1), axis=0) * 100

# Step 4: Plot
colors = ['#2ecc71', '#e74c3c']  # Green = No Alzheimer's, Red = Alzheimer's
grouped_percent.plot(kind='bar', stacked=True, color=colors, figsize=(8, 6), edgecolor='black')

# Formatting
plt.title("Alzheimer's Diagnosis by Education Group", fontsize=16)
plt.xlabel("Education Group", fontsize=14)
plt.ylabel("Percentage of Individuals", fontsize=14)
plt.legend(["No Alzheimer's", "Alzheimer's"], loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0)
plt.ylim(0, 100)

# Step 5: Annotate bars
for i, (idx, row) in enumerate(grouped_percent.iterrows()):
    plt.text(i - 0.15, row[0] / 2, f"{row[0]:.1f}%", color='white', fontweight='bold')
    plt.text(i - 0.15, row[0] + row[1] / 2, f"{row[1]:.1f}%", color='white', fontweight='bold')

plt.tight_layout()
plt.show()

```
